```python
from states import states
import pandas as pd
import json
```

```python
df_winner = pd.read_csv("winner.csv")
states_list = list(df_winner['state_abbr'])
df_winner = df_winner[df_winner.state_abbr != 'VI'].reset_index()
df_winner = df_winner.drop('index', 1)
df_winner
```

```python
final_geojson = {"type": "FeatureCollection","features":[]}
for state in states_list:
#     if state == 'VT': continue
    data = json.load(open("us_territories_geojson/{}/shape.geojson".format(state)))
    dict_add = {"type":"Feature",
                "properties":{"state_name":"{}".format(states[state]), 
                              "num_of_jp_win":"{}".format(df_winner[df_winner['state_abbr']==state]['jackpot_win_per_state'].item()), 
                              "jackpot_win_amount":"{}".format(df_winner[df_winner['state_abbr']==state]['state_jack_winnings'].item()),
                              "total_win_amount":"{}".format(df_winner[df_winner['state_abbr']==state]['state_winnings'].item()),
                              "total_draw_sales": "{}".format(df_winner[df_winner['state_abbr']==state]['state_draw_sales'].item()),
                              "total_draw_sold": "{}".format(df_winner[df_winner['state_abbr']==state]['ticket_sales'].item())},
                "geometry": data
               }
    print(states[state] + " done")
    final_geojson['features'].append(dict_add)
```
